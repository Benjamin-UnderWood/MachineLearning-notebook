# 梯度下降法

寻找最低点

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义损失函数，找到最低点？
plot_x = np.linspace(-1, 6, 140)
plot_y = (plot_x - 2.5)**2 - 1
```

```python
plt.plot(plot_x, plot_y)
```

```python
# 导数
def dJ(theta):
    return 2*(theta-2.5)
```

```python
# 损失函数
def J(theta):
    return 2*(theta-2.5)**2-1
```

```python
eta = 0.1 # 学习率，步长
epsilon = 1e-8
theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient # 向导数的负方向移1步
    
    # 是不是最小值的点，导数等于0 的点 --- 可能永远达不到这个精度
    # 每一次损失函数都要小一点
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

print(theta)
print(J(theta))  # 截距

# ≈(2.5, -1)
```

## 查看损失函数

```python
eta = 0.1 # 学习率，步长
epsilon = 1e-8
theta = -0.5
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient # 向导数的负方向移1步
    theta_history.append(theta)
    
    # 是不是最小值的点，导数等于0 的点 --- 可能永远达不到这个精度
    # 每一次损失函数都要小一点
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+') 
```

刚开始挪动幅度比较大，斜率比较陡！

后面梯度平缓，乘的比较小的数

```python
len(theta_history)
```

```python
def gradient_descent(initial_theta, eta, epsilon=1e-8):
    theta = initial_theta
    theta_history.append(initial_theta)
    
    while True:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta * gradient # 向导数的负方向移1步
        theta_history.append(theta)
    
        # 是不是最小值的点，导数等于0 的点 --- 可能永远达不到这个精度
        # 每一次损失函数都要小一点
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break

def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
```

```python
eta = 0.01
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
```
